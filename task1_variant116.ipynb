### Необходимо решить линейное уравнение переноса
$$
\frac{\partial u(x)}{\partial t} + c\frac{\partial u(x)}{\partial x}=0 \\
c = const > 0 
$$
с начальным условием при М=2

$$
\begin{cases} 
    u(0, x) = 1, x > 0 \\
    u(0, x) = exp[-(\frac{x}{\delta})^M], x \leqslant 0
\end{cases}$$

```python
from math import pow, log, sqrt, sin, cos, e, pi, exp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import animation, rc
from IPython.display import HTML

M=2 #гауссова функция'M': 4, 'c': 0.5, 'd': 2, 'a': -4, 'b': 10, 'N': 400, 'tau': 0.05, 'T': 30
c = 0.5
delta = 2
tau=0.05
T=30
```

## Неявный уголок
$$
\hat{u}_j - u_j + r(\hat{u}_j - \hat{u}_{j-1}) = 0, r = \frac{c\tau}{h}
$$

Запишем начальные условия и формулу для точного решения

```python
def phi(x, t, delta, M, c):
    #if (x > 0):
    #    return 1
    #else: 
    #    return exp(-pow((x-c*t)/delta, M))
    return exp(-pow((x-c*t)/delta, M))
    
def phi_start(x, t, delta, M, c):
    if (x-c*t > 0):
        return 1
    else: 
        return exp(-pow((x-c*t)/delta, M))
```

Равномерная сетка

```python
a = -4.0
b = 30
N = 2000
T=50
h = (b-a)/N  # step x
t = 0.05     # step t
r = c*tau/h
print('Kurand', r)
x=[]
lower=[]
results=[]

for i in range(1, N+1):
    x.append(a+h*i) #сетка
    lower.append(phi_start(x[i-1], 0, delta, M, c)) #начальные условия
print(len(x), len(lower), a+h*i)
```

считаем следующие слои

```python
t=0
while t<T:
    results.append(lower) #write old layer in results
    #start forming new layer
    layer=[phi(x[0], 0, delta, M, c)] #for i=0
    #print(len(lower))
    for i in range(1, N):
        layer.append(float(lower[i]+r*layer[i-1])/(1.0+r))
    #print(len(layer))
    lower=layer
    t+=tau
```

построим график

```python
#our ugolock
for i in range(len(results)):
    if (i%100)==0:
        plt.plot(x, results[i])

#theory
theory=[]
for i in range(len(results)):
    th1=[]
    for elem in x:
        th1.append(phi_start(elem, i*tau, delta, M, c))
    theory.append(th1)
    
for i in range(len(results)):
    if (i%100)==0:
        plt.plot(x, theory[i], color='black')
plt.show()
```

посчитаем ошибки

```python
def find_emax(results, theory, i):
    emax=0
    for j in range(len(results[i])):
        if emax<abs(results[i][j]-theory[i][j]):
            emax=abs(results[i][j]-theory[i][j])
    return emax
    
def find_erms(results, theory, i):
    erms=0
    for j in range(len(results[i])):
        foo=abs(results[i][j]-theory[i][j])
        erms+=pow(foo, 2)
    return sqrt(erms/len(results[i]))

def find_TV(results, i):
    TV=0
    for j in range(len(results[i])-1):
        TV+=abs(results[i][j+1]-results[i][j])
    return TV
```

```python
#ошибки
print 'time moment', 300*tau, ' from ', len(results)*tau
print 'emax', find_emax(results, theory, i=300) #i - time mark
print 'erms', find_erms(results, theory, i=300) #i - time mark
print 'TV', find_TV(results, i=300)

print 'time moment', 590*tau, ' from ', len(results)*tau
print 'emax', find_emax(results, theory, i=590) #i - time mark
print 'erms', find_erms(results, theory, i=590) #i - time mark
print 'TV', find_TV(results, i=590)
```

## схема Лакса-Вендроффа

$$
\large \frac{\hat{u}_j - u_j}{\tau} + \frac{\bar{f}_{j+1/2} - \bar{f}_{j-1/2}}{h} = 0
$$

считаем

```python
a = -4.0
b = 10.0
N = 500
T=30
h = (b-a)/N  # step x
t = 0.05     # step t
r = c*tau/h
delta=0.1
c=0.5
print('Kurand', r)
x=[]
lower=[]
results=[]

for i in range(1, N+1):
    x.append(a+h*i) #сетка
    lower.append(phi_start(x[i-1], 0, delta, M, c)) #начальные условия
print(len(x), len(lower), a+h*i)

def fplus(func,index):
    if (i+1==N):
        return func[index]+(func[0]-func[index])*(1-r)/2 
    return func[index]+(func[index+1]-func[index])*(1-r)/2
def fminus(func,index):
    return func[index-1]+(func[index]-func[index-1])*(1-r)/2 
```

```python
#calculate next layer
t=0
while t<T:
    results.append(lower) #write old layer in results
    #start forming new layer
    layer=[phi(x[0], 0, delta, M, c)] #for i=0
    #print(len(lower))
    for i in range(1, N):
        layer.append(-float((fplus(lower, i)-fminus(lower, i))*c*tau/h)+lower[i])
    #print(len(layer))
    lower=layer
    t+=tau
```

строим график

```python
#our scheme
for i in range(len(results)):
    if (i%100)==0:
        plt.plot(x, results[i])

#theory
theory=[]
for i in range(len(results)):
    th1=[]
    for elem in x:
        th1.append(phi_start(elem, i*tau, delta, M, c))
    theory.append(th1)
    
for i in range(len(results)):
    if (i%100)==0:
        plt.plot(x, theory[i], color='black')
plt.show()
```

```python
#ошибки
print 'time moment', 300*tau, ' from ', len(results)*tau
print 'emax', find_emax(results, theory, i=300) #i - time mark
print 'erms', find_erms(results, theory, i=300) #i - time mark
print 'TV', find_TV(results, i=300)

print 'time moment', 590*tau, ' from ', len(results)*tau
print 'emax', find_emax(results, theory, i=590) #i - time mark
print 'erms', find_erms(results, theory, i=590) #i - time mark
print 'TV', find_TV(results, i=590)
```

# КАБАРЕ

$$
\hat{u}_j - u_j + u_{j-1} - \check{u}_{j-1} + 2r(u_j - u_{j-1}) = 0, r = \frac{c\tau}{h}
$$

```python
a = -5.0
b = 10.0
N = 1000
T=30
c=0.15
h = (b-a)/N  # step x
t = 0.05     # step t
r = c*tau/h
delta=2
print('Kurand', r)
x=[]
lower=[]
results=[]

for i in range(1, N+1):
    x.append(a+h*i) #сетка
    lower.append(phi_start(x[i-1], 0, delta, M, c)) #начальные условия
deep=lower
print(len(x), len(lower), a+h*i)
```

```python
#calculate next layer
t=0
while t<T:
    results.append(lower) #write old layer in results
    #start forming new layer
    layer=[phi(x[0], 0, delta, M, c)] #for i=0
    #print(len(lower))
    for i in range(1, N):
        layer.append( lower[i]-lower[i-1]+deep[i-1]-2*r*(lower[i]-lower[i-1]) )
    #print(len(layer))
    deep=lower
    lower=layer
    t+=tau
```

```python
#our scheme
for i in range(len(results)):
    if (i%100)==0:
        plt.plot(x, results[i])

#theory
theory=[]
for i in range(len(results)):
    th1=[]
    for elem in x:
        th1.append(phi_start(elem, i*tau, delta, M, c))
    theory.append(th1)
    
for i in range(len(results)):
    if (i%100)==0:
        plt.plot(x, theory[i], color='black')
plt.show()
```
